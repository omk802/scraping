```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
url = "https://eur-lex.europa.eu/legal-content/EN/TXT/HTML/?uri=CELEX:02014D0145-20220916&qid=1665724867900&from=en"
data = requests.get(url).text
# print(data)
```

```python
soup = BeautifulSoup(data, 'html.parser')
val = []
```

```python
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
    val.append(table.get('class'))
```

```python
d = {"no": [],"name":[],"info":[]} 

for i in val: 
    if i is not None:
        tables = soup.find_all("table")
#         print(tables)

        for row in tables[3].tbody.find_all('tr'): 
                           
#                
                co = row.find_all(attrs={'valign': 'top'})
                if (len(co) == 5):
                      no = co[0].text.replace("\n"," ")
                      name = co[1].text
                      info = co[2].text
                
                      d['no'].append(no)
                      d['name'].append(name)
                      d['info'].append(info)
#                       
                     
        for r in tables[4].tbody.find_all('tr'): 
                           
#                
                    co = r.find_all(attrs={'valign': 'top'})
                    if (len(co) == 5):
                          no = co[0].text.replace("\n"," ")
                          name = co[1].text
                          info = co[2].text
                          d['no'].append(no)
                          d['name'].append(name)
                          d['info'].append(info)    
                   
        break
  


df = pd.DataFrame.from_dict(d)
# movie = pd.DataFrame({'gender':gender})
# print(movie)
df.to_csv("om.csv")
pd.read_csv('om.csv')
```

```python
def get_gender(ls):
    ls = [i.strip() for i in ls]
    # check for substr Gender
    rs = [i for i in ls if 'Gender' in i]

    # check len of list to check if gender exist else return nan
    if len(rs) > 0:
        return rs[0].split(':')[1].strip()
    else:
        return np.nan
def get_dob(ls):
    ls = [i.strip() for i in ls]

    # check for substr Gender
    rs = [i for i in ls if 'DOB' in i]

    # check len of list to check if gender exist else return nan
    if len(rs) > 0:
        return rs[0].split(':')[1].strip()
    else:
        return np.nan
def get_pob(ls):
    ls = [i.strip() for i in ls]

    # check for substr Gender
    rs = [i for i in ls if 'POB' in i]

    # check len of list to check if gender exist else return nan
    if len(rs) > 0:
        return rs[0].split(':')[1].strip()
    else:
        return np.nan
def get_nation(ls):
    ls = [i.strip() for i in ls]

    # check for substr Gender
    rs = [i for i in ls if 'Nationality' in i]

    # check len of list to check if gender exist else return nan
    if len(rs) > 0:
        return rs[0].split(':')[1].strip()
    else:
        return np.nan
def get_address(ls):
    ls = [i.strip() for i in ls]

    # check for substr Gender
    rs = [i for i in ls if 'Address' in i]

    # check len of list to check if gender exist else return nan
    if len(rs) > 0:
        return rs[0].split(':')[1].strip()
    else:
        return np.nan
def get_name(ls):
    ls = [i.strip() for i in ls]
#     print(ls)

    # check for substr Gender
    rs = [i for i in ls]
#     print(rs)
    # check len of list to check if gender exist else return nan
    if len(rs) > 0:
        return rs[1].strip()
    else:
         return np.nan
```

```python
df['Name'] = df['name'].apply(lambda x:get_name(x.split('\n')))
df['Gender'] = df['info'].apply(lambda x:get_gender(x.split('\n')))
df['DOB'] = df['info'].apply(lambda x:get_dob(x.split('\n')))
df['POB'] = df['info'].apply(lambda x:get_pob(x.split('\n')))
df['Nationality'] = df['info'].apply(lambda x:get_nation(x.split('\n')))
df['Address'] = df['info'].apply(lambda x:get_address(x.split('\n')))
# df['Name'] = df['name'].apply(lambda x:get_address(x.split('\n')))
```

```python
df.drop('info', axis=1, inplace=True)
df.drop("name", axis =1 , inplace = True)
```
